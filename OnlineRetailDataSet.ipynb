```python
def biggerOrderItems(x):
    if(x["TotalnoOfItems"]>500):
        return 'Yes'
    else:
        return 'No'

    
    
def Loyal(x):
    if(x["NoOfDays"]>100):
        return True
    else:
        return False


def Active(x):
    if(x["noOfDaysActiveBefore"]<2000):
        return True
    else:
        return False
def Cancel(x):
    if(x["cancel"]==0):
        return 'No'
    else: 
        return 'Yes'
```

```python
New_table=newt
New_table["Bigger_Order&Items"]=New_table.apply(biggerOrderItems,axis=1)
New_table["Loyal"]=New_table.apply(Loyal,axis=1)
New_table["Active"]=New_table.apply(Active,axis=1)
New_table["CancellingUser"]=New_table.apply(Cancel,axis=1)
```
